## II. Visuals

### <b> Problem 2 </b> &nbsp; Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
# Calculate average grades by Gender
avg_by_gender = df.groupby('Gender')['Average'].mean().reset_index()

# Calculate average grades by Track
avg_by_track = df.groupby('Track')['Average'].mean().reset_index()

# Calculate average grades by Hometown
avg_by_hometown = df.groupby('Hometown')['Average'].mean().reset_index()
```

```python
# Plotting
plt.figure(figsize = (14, 6))

# Plot average grades by Gender
plt.subplot(1, 3, 1)
sns.barplot(x = 'Gender', y = 'Average', data = avg_by_gender)
plt.title('Average Grade by Gender')

# Plot average grades by Track
plt.subplot(1, 3, 2)
sns.barplot(x = 'Track', y = 'Average', data = avg_by_track)
plt.title('Average Grade by Track')

# Plot average grades by Hometown
plt.subplot(1, 3, 3)
sns.barplot(x = 'Hometown', y = 'Average', data = avg_by_hometown)
plt.title('Average Grade by Hometown')

plt.tight_layout()
plt.show()
```
